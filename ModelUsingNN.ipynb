```python
from keras.models import Sequential
from keras.layers import Dense 
import pandas as pd 
```

```python
#data loading 
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
# data cleaning
data.info()
# our data is clean ( all values are numericul and we dont have the nan and null values )
```

```python
# data spliting 
from sklearn.model_selection import train_test_split
x = data.drop(["Outcome"] , axis = 1 )
y = data['Outcome']
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size= 0.3 )
```

```python
# building model 
from keras.optimizers import Adam
model = Sequential()
model.add(Dense(10 , input_dim = x_train.shape[1] , activation="relu"))
model.add(Dense(4 , activation="relu"))
model.add(Dense(1 , activation= "sigmoid"))
model.compile(optimizer="adam" , loss='binary_crossentropy' , metrics=['accuracy'])
model.fit(x_train, y_train , epochs=500)
```

```python
# model evaluation 
from sklearn.metrics import confusion_matrix , accuracy_score 
y_pred = (model.predict(x_test) > 0.5 ).astype(int)
print(y_pred)
print("\n the accuracy score of our model " , accuracy_score(y_test , y_pred ))
```

```python
# compare the outputs 
nbr = 0 
for i in range(x_test.shape[0]) : 
    if y_test.iloc[i] == y_pred[i] : 
        nbr = nbr + 1 
print(f"the true values {nbr} in {x_test.shape[0]} values ")
```
